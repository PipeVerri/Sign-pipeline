```python
from faster_whisper import WhisperModel
```

```python
model = WhisperModel("large-v3-turbo", device="cuda")
```

```python
def format_timestamp(seconds):
    hours = int(seconds // 3600)
    minutes = int((seconds % 3600) // 60)
    secs = int(seconds % 60)
    millis = int((seconds % 1) * 1000)
    return f"{hours:02d}:{minutes:02d}:{secs:02d}.{millis:03d}"
```

```python
def process_audio(filename, path):
    segments, info = model.transcribe(path + filename, word_timestamps=True, beam_size=5, language="es")
    with open(f"/kaggle/working/{filename.split('.')[0]}.vtt", "w", encoding="utf-8") as f:
        f.write("WEBVTT\n\n")
    
        for i, segment in enumerate(segments, start=1):
            start = format_timestamp(segment.start)
            end = format_timestamp(segment.end)
            text = segment.text.strip()
        
            f.write(f"{start} --> {end}\n")
            f.write(f"{text}\n\n")
```

```python
import os
from tqdm.notebook import tqdm

path = "/kaggle/input/lsa-x-audio/audio/"
files = os.listdir(path)

for f in tqdm(files, desc="Processing audio files"):
    process_audio(f, path)
```
