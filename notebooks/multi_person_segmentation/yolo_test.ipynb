```python
import cv2

cap = cv2.VideoCapture("../../data/raw/0-AsociacionCivil/0-300.mp4")
# Set start frame
fps = cap.get(cv2.CAP_PROP_FPS)
start_frame = int(120 * fps)  # 2 minutes * 60 seconds * fps
cap.set(cv2.CAP_PROP_POS_FRAMES, start_frame)
```

```python
from ultralytics import YOLO
from src.shared.utils.video import frame_reader

model = YOLO("../../models/yolo/yolo11s.pt")

for frame in frame_reader(cap, fps=6):
    results = model.track(
        frame,
        persist=True,
        classes=[0] # Only track humans
    )
    print(results)
    break

    img = results[0].plot()
    cv2.imshow("tracking", img)
    cv2.waitKey(1)
```
